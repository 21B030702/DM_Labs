```python
#Exercise 1: Implementing Basic Classification Algorithms

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# 1. Загрузка данных Iris
iris = load_iris()
X, y = iris.data, iris.target

# 2. Разделение данных на обучающую и тестовую выборки
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 3. Логистическая регрессия
logreg = LogisticRegression(max_iter=200)
logreg.fit(X_train, y_train)
y_pred_logreg = logreg.predict(X_test)

# 4. K-ближайших соседей (KNN)
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)

# 5. Оценка точности моделей
logreg_accuracy = accuracy_score(y_test, y_pred_logreg)
knn_accuracy = accuracy_score(y_test, y_pred_knn)

print(f"Точность логистической регрессии: {logreg_accuracy:.2f}")
print(f"Точность KNN: {knn_accuracy:.2f}")
```

```python
#Exersice 2: Confusion Matrix and Classification Report
from sklearn.metrics import confusion_matrix, classification_report

# Матрица ошибок и отчет для логистической регрессии
print("Логистическая Регрессия - Матрица Ошибок:")
print(confusion_matrix(y_test, y_pred_logreg))
print("Логистическая Регрессия - Отчет по Классификации:")
print(classification_report(y_test, y_pred_logreg))

# Матрица ошибок и отчет для KNN
print("KNN - Матрица Ошибок:")
print(confusion_matrix(y_test, y_pred_knn))
print("KNN - Отчет по Классификации:")
print(classification_report(y_test, y_pred_knn))
```

```python
#Exercise 3: Implementing K-Means Clustering
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# 1. Создание тестового датасета для кластеризации
X, _ = make_blobs(n_samples=300, centers=4, random_state=42)

# 2. Обучение K-Means
kmeans = KMeans(n_clusters=4, random_state=42)
y_kmeans = kmeans.fit_predict(X)

# 3. Визуализация кластеров
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', marker='x')
plt.title("K-Means Clustering")
plt.show()
```

```python
#Exercise 4: Evaluating K-Means Clustering
inertia = []
K = range(1, 10)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

# Визуализация метода локтя
plt.plot(K, inertia, 'bo-')
plt.xlabel('Количество кластеров')
plt.ylabel('Inertia')
plt.title('Метод Локтя')
plt.show()
```

```python
#Exercise 5: Implementing Hierarchical Clustering
from scipy.cluster.hierarchy import dendrogram, linkage

# 1. Применение иерархической кластеризации
Z = linkage(X, method='ward')

# 2. Визуализация дендрограммы
plt.figure(figsize=(10, 7))
dendrogram(Z)
plt.title("Dendrogram for Hierarchical Clustering")
plt.show()
```

```python
#Exercise 6: Introduction to DBSCAN
from sklearn.cluster import DBSCAN

# 1. Применение DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
y_dbscan = dbscan.fit_predict(X)

# 2. Визуализация кластеров
plt.scatter(X[:, 0], X[:, 1], c=y_dbscan, cmap='plasma')
plt.title("DBSCAN Clustering")
plt.show()
```
